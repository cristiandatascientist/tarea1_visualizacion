### Descripción de Columnas:

* UserID: Un identificador único asignado a cada usuario.

## Ingeniería de Carasteristicas 


Para realizar Codificación de variables categóricas y Normalización o Estandarización vamos a aplicar lo siguiente:

Codificación de variables categóricas: Vamos a utilizar pd.get_dummies() para las variables categóricas y convertirlas en variables dummy o variables de tipo binario.

Normalización o Estandarización: Aplicaremos StandardScaler o MinMaxScaler de sklearn dependiendo si se desea estandarizar o normalizar datos.

### Codificación de Variables Categóricas

Las variables categóricas, como Gender, Location, y Platform, etc., no pueden ser interpretadas directamente por los modelos de ML, ya que trabajan con variables de tipo continua. La codificación transforma estas variables categóricas en variables binarias o dummies, permitiendo que los modelos las utilicen para hacer predicciones.

Es necesario realizar la codificación de las variables categoricas para todas las columnas porque los modelos no pueden procesar texto o categorías nominales. Cada columna categórica debe ser transformada en un formato numérico, lo que es clave para el análisis y modelado.

```python
# Codificación de variables categóricas
df_encoded = pd.get_dummies(df_redes_sociales, columns=['Gender', 'Location', 'Profession', 'Demographics', 'Platform', 
                                                        'Video Category', 'Frequency', 'Watch Reason', 
                                                        'DeviceType', 'OS', 'CurrentActivity', 'ConnectionType'])

# Ver las primeras filas del dataframe codificado
print(df_encoded.head())
```

### Visualización de la distribución de una de las variables categóricas Gender.

```python
import pandas as pd
from plotnine import ggplot, aes, geom_bar, labs, theme_classic, scale_fill_manual

# Corregir dataframe: Sumar las columnas de género y crear una columna 'Gender'
gender_counts = pd.DataFrame({
    'Gender': ['Female', 'Male'],
    'Count': [df_encoded['Gender_Female'].sum(), df_encoded['Gender_Male'].sum()]
})

# Crear la gráfica de barras con plotnine
plot = (
    ggplot(gender_counts, aes(x='Gender', y='Count', fill='Gender')) +
    geom_bar(stat='identity') +
    labs(title='Distribución de Géneros Codificados', y='Cantidad de Usuarios', x='Género') +
    scale_fill_manual(values={'Female': 'skyblue', 'Male': 'orange'}) +
    theme_classic()  # Cambiar el tema a uno más claro
)

# Mostrar la gráfica
plot.show()
```

### Normalización y/o Estandarización de Variables Numéricas


La normalización es útil cuando las características tienen escalas muy diferentes. Esto permite que todas las variables numéricas estén en la misma escala (entre 0 y 1). Es crucial cuando los modelos como regresiones o algoritmos basados en distancias (K-NN, SVM) se ven afectados por diferencias en la magnitud de las variables.

Es necesario hacerlo para las columnas numéricas: Especialmente en este caso, donde las columnas como Income y Age pueden tener rangos muy diferentes, aplicar normalización asegura que las variables tengan el mismo peso durante el análisis.

```python
from sklearn.preprocessing import MinMaxScaler

# Seleccionar columnas numéricas para normalizar
numerical_columns = ['Age', 'Income', 'Total Time Spent', 'Number of Sessions', 'Video Length', 
                    'Engagement', 'Importance Score', 'Time Spent On Video', 'Number of Videos Watched', 
                    'Scroll Rate', 'ProductivityLoss', 'Satisfaction', 'Self Control', 'Addiction Level']

# Aplicar MinMaxScaler
scaler = MinMaxScaler()
df_encoded[numerical_columns] = scaler.fit_transform(df_encoded[numerical_columns])

# Ver los primeros datos normalizados
print(df_encoded.head())
```

### Histograma de la variable "Ingreso" antes y después del proceso de normalización.

```python
# Visualización antes y después de la normalización
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Antes de la normalización
df_redes_sociales['Income'].plot(kind='hist', ax=ax[0], color='skyblue')
ax[0].set_title('Distribución de Ingresos Antes de Normalización')

# Después de la normalización
df_encoded['Income'].plot(kind='hist', ax=ax[1], color='orange')
ax[1].set_title('Distribución de Ingresos Después de Normalización')

plt.show()
```

### Estandarización (Media 0 y desviación estándar 1):

La Estandarización es útil cuando necesitamos que las características tengan media 0 y desviación estándar 1. Esto es especialmente importante en algoritmos como las redes neuronales, que pueden ser sensibles a las diferentes escalas de las variables.

Es necesario hacerlo para las columnas numéricas, especialmente si se usa un modelo que suponga que los datos están distribuidos normalmente (como regresión logística o regresión lineal).

```python
from sklearn.preprocessing import StandardScaler

# Aplicar StandardScaler para estandarización
scaler = StandardScaler()
df_encoded[numerical_columns] = scaler.fit_transform(df_encoded[numerical_columns])

# Ver los primeros datos estandarizados
print(df_encoded.head())
```

#### Visualización: El cambio en la media y desviación estándar de una variable numérica "Satisfaction" antes y después de la estandarización

```python
# Visualización antes y después de la estandarización
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Antes de la estandarización
df_redes_sociales['Satisfaction'].plot(kind='hist', ax=ax[0], color='skyblue')
ax[0].set_title('Distribución de Satisfacción Antes de Estandarización')

# Después de la estandarización
df_encoded['Satisfaction'].plot(kind='hist', ax=ax[1], color='orange')
ax[1].set_title('Distribución de Satisfacción Después de Estandarización')

plt.show()
```
